```python
import csv
import requests
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
File = 'lianjia.xlsx'
df = pd.read_excel(File, encoding='utf-8',usecols=['price',  '城区'])
df.boxplot(by='城区', column='price')
plt.title('北京市二手房总价分析', fontsize=20)  # 标题，并设定字号大小
plt.suptitle('')  # 隐藏默认的groupby文字

plt.savefig('二手房.jpg')
plt.show()
```
