```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from statsmodels.stats.diagnostic import acorr_ljungbox as lb_test

import statsmodels.api as sm

import statsmodels.tsa.api as smt
```

## 畫時序圖, ACF, PACF

```python
def drawts(y,pname):
    ##draw ax
    fig = plt.figure(figsize=(10,8))
    ts_ax=plt.subplot2grid((2,2),(0,0),colspan=2)
    acf_ax=plt.subplot2grid((2,2),(1,0))
    pacf_ax=plt.subplot2grid((2,2),(1,1))
    ##draw plot
    ts_ax.plot(y,'*-')
    ts_ax.set_title('Time Series Analysis Plots')
    smt.graphics.plot_acf(y,lags=None,ax=acf_ax,alpha=0.05) ##2sigma
    smt.graphics.plot_pacf(y,lags=None,ax=pacf_ax,alpha=0.05)  ##2sigma
    #plt.savefig('%s.jpg'%pname,dpi=256)
    plt.show()
    plt.close()
```

## LBtest

```python
def purerandtest(y):
    a,b=lb_test(y,lags=None,boxpierce=False)
    LB_purerand=pd.DataFrame(np.c_[a,b],columns=['LB','Pvalue'])
    LB_purerand['lags']=range(1,len(a)+1)
    print('----time series: LB pure randomness test----')
    print(LB_purerand)
```

## 原始數據和擬合數據比較圖

```python
def plotcon(y,ypred,pname):
    plt.plot(y,'*-',label='observe')
    plt.plot(ypred,label='fitted')
    plt.legend()
    plt.title('observe and fitted')
    #plt.savefig('%s_fitted.jpg'%pname,dpi=256)
    #plt.close()
    plt.show()
```

## 開始分析

```python
##read data
dfname='data1.10'
y=pd.read_csv('%s.csv'%dfname,header=None)
y.iloc[:,0]=y.iloc[:,0].astype('float')
y=y.values[:,0]

##plot
drawts(y,dfname)
#purerandtest(y)
```

## 季節指數Sj

```python
##calculate Sij

S=np.zeros((12))
for i in range(12):
    S[i]=np.mean(y[i::12])/np.mean(y)
print('Si:')
print(S)
plt.plot(np.arange(1,12+1),S)
plt.xlabel('month')
plt.ylabel('Si')
plt.title('Si')
plt.show()

ans=pd.DataFrame(np.c_[np.arange(1,13),y.reshape(-1,12).T,y.reshape(-1,12).T.mean(1),S],columns=['月','1995','1996','1997','1998','1999','2000','月平均','季節指數'])
print(ans)
print(y.mean())
```
